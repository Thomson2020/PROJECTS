```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import spherical_jn

# Constants
hbar = 1  # Planck's constant divided by 2π (for simplicity)
m = 1     # Mass (for simplicity)

# Define the range of radii within the spherical well
r = np.linspace(0, 14, 500)  # Adjust the range as needed

# Define the maximum value of n for energy states
n_max = 3  # You can change this to get more energy states

# Create a function to solve the radial Schrödinger equation
def solve_radial_schrodinger(n, l):
    k_n = n * np.pi / r[-1]  # Wavenumber for the n-th energy level
    psi_nl = spherical_jn(l, k_n * r)  # Spherical Bessel function
    energy = hbar**2 * k_n**2 / (2 * m)
    return psi_nl, energy

# Plot energy states for the spherical well
plt.figure(figsize=(8, 6))
for n in range(1, n_max + 1):
    for l in range(n):
        psi_nl, energy = solve_radial_schrodinger(n, l)
        plt.plot(r, psi_nl, label=f'n={n}, l={l}, E={energy:.4f}')

plt.xlabel('Radius (r)')
plt.ylabel('Ψ(r)')
plt.xlim(0,14)
plt.ylim((-1,1))
plt.title('Radial Schrödinger Equation for Spherical Infinite Well')
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()
```
